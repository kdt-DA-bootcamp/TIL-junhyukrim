```python
import requests
from bs4 import BeautifulSoup
URL = 'https://yozm.wishket.com/magazine/list/develop/?page=2'
response = requests.get(URL)
soup = BeautifulSoup(response.text, 'html.parser')
result = soup.find('title')
print(result)
print(result.text)
```

```python
title_tag = soup.find('a', class_='item-title')
print(title_tag.text)

#속성 추출
title_tag.attrs['class']
title_tag.attrs['href']
```

```python
tags = soup.find_all('a', class_='item-title')
print(tags)

for tag in tags:
    print(tag.text)
    print(tag.attrs['href'])

print(tags[0].text, tags[0].attrs['href'])
```
